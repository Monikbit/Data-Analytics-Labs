```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('learningSet.csv')
```

```python
data.head()
```

```python
def clean_headers(x):
  for i in x.columns:
   x.columns = [i.lower().replace(' ','_')  for i in x.columns]
  return x

clean_headers(data)
```

```python
# Separamos las variables numericas y categoricas
categorical=data.select_dtypes(include=["object"])
numerical=data.select_dtypes(include=["number"])
```

```python
categorical.replace(' ', np.nan, inplace=True)
categorical.replace('  ', np.nan, inplace=True)
categorical.replace('   ', np.nan, inplace=True)
categorical.replace('    ', np.nan, inplace=True)
```

```python
# Check and deal with NaN values.
print("Nulos")
for i in categorical.columns: 
  
  print( i, 100*sum(categorical[i].isnull()) / len(data[i]) ,"%")
```

```python
# Creamos una lista vacia
drop_list=[]
len(drop_list)
```

```python

drop_list=['osource', 'zip']
drop_list
```

```python
# Creamos una datframe con los nombres de las columnas de nuestra data original y el porcentaje de nulos
nulls_columns=pd.DataFrame(categorical.isna().sum()/len(categorical)).reset_index()
nulls_columns.columns=['column_name','null_percentage']
nulls_columns
```

```python
# Creamos un dataframe que contenga nuestros nombres de columnas, pero que cumplan con un porcentaje de nulos mayor a 85%
columns_above_threshold=nulls_columns[nulls_columns['null_percentage']>0.85]
print(len(columns_above_threshold))
columns_above_threshold
```

```python
drop_columns_list=list(columns_above_threshold['column_name'])
drop_columns_list
```

```python
# Unimos los valores de esa lista, con la lista que creamos arriba simplemente sumando las listas
# drop_columns_list=drop_columns_list + drop_list
# drop_columns_list
```

```python
# con append
for x in drop_list:
    drop_columns_list.append(x)
drop_columns_list
```

```python
data=data.drop(drop_columns_list, axis=1)
data
```

```python
categorical=data.select_dtypes(include=["object"])
numerical=data.select_dtypes(include=["number"])
```

```python
# - Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest
# - Note that there are a few null values in the column. We will first replace those null values using the code below:

print(data['gender'].value_counts())
data['gender'].replace(' ', np.nan, inplace=True) # Vamos a reemplazar los espacios con nulos
data['gender'] = data['gender'].fillna('F')  # Vamos a llenar los valores nulos con F
```

```python
data['gender'].isnull().sum()
```

```python
data['gender'].value_counts()
```

```python
# reducimos los valores
data['gender']= data['gender'].replace({'F': 'F', 'M': 'M', 'U': 'Other', 'J': 'Other', 'C': 'Other', 'A': 'Other'})
data['gender'].value_counts()
```
